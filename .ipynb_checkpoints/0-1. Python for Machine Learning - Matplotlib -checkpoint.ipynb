# Python for Machine Learning - Matplotlib

Matplotlib은 플로팅 라이브러리입니다. **Matplotlib**은 쉬운 일을 쉽게, 어려운 일은 가능하게 만들어 줍니다. 단지 몇 줄의 코드로 플롯, 히스토그램, 파워 스펙트럼, 막대 차트, 오류 차트, 스캐터플롯 등을 생성 할 수 있습니다.

라이브러리 설명서 : http://matplotlib.org/

이 섹션에서는 머신러닝에 유용한 플롯팅 시스템을 제공하는 matplotlib.pyplot 모듈에 대해 간략하게 소개합니다.

### Plotting
matplotlib의 가장 중요한 기능은 플롯(plot)으로 2D 데이터를 플로팅 할 수 있습니다. 다음은 간단한 예입니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3*np.pi, 0.1)   # 0 ~ 3*pi 를 0.1 간격으로 샘플링
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

추가 작업을 조금만 하면 한 번에 여러 줄을 쉽게 그릴 수 있고, 제목, 범례 및 축 레이블을 추가 할 수 있습니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine using pyplot()')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

[여기](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)에서 plot 함수에 대해 더 많이 읽을 수 있습니다.
또는, 다음과 같이 subplots()을 사용하여 같은 플로팅을 할 수도 있다.

```python
fig, ax = plt.subplots()
ax.plot(x, y_sin, x, y_cos)
ax.set_title("Sine and Cosine using pyplot.subplots()")
ax.set_xlabel("x axis label")
ax.set_ylabel("y axis label")
ax.legend(['Sine', 'Cosine'])
plt.show()
```

### subplot
subplot 함수를 사용하여 같은 그림에 다른 것들을 그릴 수 있습니다. 다음은 그 예입니다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x, y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))

x = np.linspace(0, 5, 10)
axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].grid()
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_xlim([2, 5])
axes[2].set_ylim([0, 60])
axes[2].set_title("custom axis range")
plt.show()
```

[여기](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplots)에서 subplots 함수에 대한 자세한 내용을 볼 수 있습니다.

### Bar

```python
import matplotlib.pyplot as plt

x = [2, 4, 6, 8, 10]
y = [6, 7, 8, 2, 4]

x2 = [1, 3, 5, 7, 9]
y2 = [7, 8, 2, 4, 2]

plt.bar(x, y, label="Bars 1", color="r")
plt.bar(x2, y2, label="Bars 2", color="g")

plt.show()
```

### Scatter plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 2, 1, 4, 5, 2]

plt.scatter(x, y, label="scatter plot", color="b", marker="*")
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.legend()
plt.show()
```

### 이미지
imshow 함수를 사용하여 이미지를 표시 할 수 있습니다. 다음은 그 예입니다.

```python
import numpy as np

img = plt.imread('cat.jpg')
print(img.dtype, img.shape)

# We can tint the image by scaling each of the color channels
# by a different scalar constant. The image has shape (400, 248, 3);
# we multiply it by the array [1, 0.95, 0.9] of shape (3,);
# numpy broadcasting means that this leaves the red channel unchanged,
# and multiplies the green and blue channels by 0.95 and 0.9
# respectively.
img_tinted = img * [1, 0.95, 0.9]
print(img_tinted.dtype, img_tinted.shape)

plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(np.uint8(img_tinted))
plt.show()
```
